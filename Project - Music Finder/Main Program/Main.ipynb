```python
#This function allows the 
def find_song_by_lyrics(snippet, song_data):
    snippet = snippet.lower()

    matches = []

    for url, file_name in song_data.items():
        with open(file_name, "r", encoding="utf-8") as f:
            lyrics = f.read().lower()

            if snippet in lyrics:
                matches.append((url, file_name))

    return matches
```
