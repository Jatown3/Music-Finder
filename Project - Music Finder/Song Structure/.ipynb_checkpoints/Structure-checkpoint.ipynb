```python
import lyricsgenius
import billboard
import json
import schedule
import time
import os

# Reading file to get access token for Genius API
with open("access_token.txt", "r") as file:
    access_token = file.read().strip()
    
# Creating genius object
genius = lyricsgenius.Genius(access_token)

# Run program to get final song data and set a schedule for every day monday at 10:00 AM
run_pipeline()
job = schedule.every().monday.at("10:00").do(run_pipeline)
print("Next run:", job.next_run)

running = True

# Runs schedule and program forever and checks every 60 seconds if it's time to run pipeline again
while running:
    schedule.run_pending()
    time.sleep(60)
```

```python
# Runs program and stores final data structure into a json file for later manipulation
def run_pipeline():
    print("Initializing pipeline...")
    
    deleting_old_songs()
    
    songs = get_trending_songs()
    
    data = process_songs(songs)

    with open ("data.json","w") as file:
        json.dump(data, file, indent=4)

    print("Song data saved to data.json")
    print("Pipeline finished")

    
```

```python
# Processes trending songs and writes lyrics file and info file for each song, storing it into a dict
# Data Structure: {"URL of song": {"Lyrics": song_lyrics file, "Info": song_info file}}
def process_songs(songs):
    song_folder = "List_of_Songs"
    info_folder = "Song_Info"

    song_data = {}

    # Processing all the songs from trending song data from billboard
    print("Processing Songs...")
    for i, song in enumerate(songs, start = 1):
        song_information = {}

        try:
            result = genius.search_song(song["Title"], song["Artist"])

            # Filter's song's by comparing genius result to trending song data from billboard, making sure it's correct song
            if song["Title"].lower() in result.title.lower() or song["Artist"].lower() in result.artist.lower():
                lyrics_file = f"{song_folder}/song{i}.txt"
                info_file = f"{info_folder}/song{i}_info.txt"

                # Writes song lyrics to lyrics file
                song_lyrics = result.lyrics
                url = result.url
                with open(lyrics_file, "w", encoding = "utf-8") as file:
                    file.write(song_lyrics)

                # Gets song album data if applicable
                album_data = result.album 
                if album_data: 
                    album_name = album_data.get("name", "N/A") 
                    album_art = album_data.get("cover_art_url", "N/A") 
                    release_date = album_data.get("release_date_for_display", "N/A") 
                else: 
                    album_name = "N/A" 
                    album_art = "N/A" 
                    release_date = "N/A"

                # Writes song information to info file
                with open(info_file, "w", encoding = "utf-8") as file:
                    file.write("===Song Information===\n\n")
                    file.write(f"Song: {song['Title']}\n")
                    file.write(f"Artist: {song['Artist']}\n")
                    file.write(f"Album: {album_name}\n")
                    file.write(f"Album Art: {album_art}\n")
                    
                song_data[url] = {"Lyrics": lyrics_file, "Info": info_file}

        except Exception as e:
            print(f"Error: {song['Title']} - {e}")

    print("Song's Processing Complete!")

    return song_data
            
    
```

```python
# Deletes previous song_lyrics and song_info files in specified folders.

def deleting_old_songs():
    print("Deleting old song files...")
    
    song_count = 0
    info_count = 0
    lyrics_folder= "List_of_Songs"
    info_folder = "Song_info"

    if os.path.exists(lyrics_folder):
        for file in os.listdir(lyrics_folder):
            if file.startswith("song") and file.endswith(".txt"):
                os.remove(os.path.join(lyrics_folder, file))
                song_count += 1

    if os.path.exists(info_folder):
        for file in os.listdir(info_folder):
            if file.startswith("song") and file.endswith(".txt"):
                os.remove(os.path.join(info_folder, file))
                info_count += 1

    print("Number of song files deleted: " + str(song_count))
    print("Number of song info files deleted: " + str(info_count))
```

```python
# Obtains the Top weekly 100 songs from Billboard and appends a dict of the song's artist and title to a list.
# Data Structure: [{"Title": title of song, "Artist": song artist}]
def get_trending_songs():

    print("Obtaining Trending Songs...")
    chart = billboard.ChartData('hot-100')
    
    songs = []
    for song in chart:
        songs.append({"Title": song.title, "Artist": song.artist})

    return songs
```
