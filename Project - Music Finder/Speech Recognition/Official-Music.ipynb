```python
import re
from IPython.display import Image, display

def display_results(results):
    if not results:
        print("No songs found matching that lyric.")
        return
    print(f"\nFound {len(results)} match(es):")
    for url, file_name in results:
        with open(file_name["Info"], "r", encoding="utf-8") as f:
            info_text = f.readlines()
        for line in info_text:
            match = re.search(r"Album Art:\s*(https?://\S+)", line)
            if match:
                url = match.group(1)
                display(Image(url=url))
            else:
                print(line)

                
        print(f"  Info File : {file_name['Info']}")
        print(f"  Lyrics File: {file_name['Lyrics']}")
        print(f"  URL: {url}")
        print("\n--- Lyrics ---")
        with open(file_name["Lyrics"], "r", encoding="utf-8") as f:
            print(f.read())
 
 
def top_three(artist_name, song_data):
    artist_name = artist_name.lower()
    matches = []
    for url, file_name in song_data.items():
        with open(file_name["Info"], "r", encoding="utf-8") as f:
            info_text = f.read().lower()
            if "artist: " + artist_name in info_text:
                matches.append((url, file_name))
                if len(matches) == 3:
                    break
    return matches
 
def find_song_title(title, song_data):
    title = title.lower()
    matches = []
    for url, info in song_data.items():
        with open(info["Info"], "r", encoding="utf-8") as l:
            lyrics = l.read().lower()
            if "song: " + title in lyrics:
                matches.append((url, info))
    return matches
```

```python
import json
import speech_recognition as sr

def recognize_speech():
    recognizer = sr.Recognizer()
    with sr.Microphone() as source:
        print("Listening... speak now")
        recognizer.adjust_for_ambient_noise(source)
        audio = recognizer.listen(source)

    try:
        text = recognizer.recognize_google(audio)
        print(f"   You said: '{text}'")
        return text
    except sr.UnknownValueError:
        print("Could not understand — try again")
        return None
    except sr.RequestError:
        print("Speech service unavailable")
        return None

def spoken_to_choice(text):
    if text is None:
        return None
 
    text = text.lower().strip()
 
    mapping = {"one":   "1",
        "two":   "2",
        "three": "3",
        "four":  "4",
        
        "search by verse":    "1",
        "search by lyric":    "1",
        "search by artist":   "2",
        "top three":          "2",
        "search by title":    "3",
        "search by song":     "3",
        "quit":               "4",
        "exit":               "4",
        "done":               "4",
    }
    
    if text in mapping:
        return mapping[text]
    
    if text in ("1", "2", "3", "4"):
        return text
        
    return text

def speak_prompt(prompt_text):
    print(prompt_text)
    return recognize_speech()
```
